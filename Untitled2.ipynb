```python
def detect_outlier(col):

    outliers = []
    threshold = 3                       
    mean = np.mean(col)
    std = np.std(col)
    
    for i in col : 
        z_score = (i - mean) / std            
        if np.abs(z_score) > threshold :
            outliers.append(i)
    return outliers
```

```python
#Feature Selection
df_test=df[df['Loan_Status'].isnull()]
df_train=df[~df['Loan_Status'].isnull()]
```

```python
X=df_train.drop('Loan_Status',axis=1)
```

```python
y=df_train['Loan_Status']
```

```python
#From Logistic Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import recall_score, f1_score, accuracy_score, precision_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)

model_lr = LogisticRegression(random_state=3)

model_lr.fit(X_train, y_train)

y_pred = model_lr.predict(X_test)

print(f'The recall_score is : {recall_score(y_test, y_pred)}')  
print(f'The f1_score is : {f1_score(y_test, y_pred)}') 
print(f'The precision_score is : {precision_score(y_test, y_pred)}')  
print(f'The accuracy_score is : {accuracy_score(y_test, y_pred)}\n') 


scores = cross_val_score(model_lr, X_train, y_train, cv=5)
mean_accuracy = scores.mean()

print(f"Mean training accuracy (5-fold CV): {mean_accuracy:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV
# define the hyperparameter grid
param_grid = {'criterion': ['gini', 'entropy'], 'max_depth': [2, 3, 4, 5, 6, 7, 8]}

# create a GridSearchCV object
grid_search = GridSearchCV(model_rf, param_grid, cv=5)

# fit the GridSearchCV object to the data
grid_search.fit(X, y)

# print the best hyperparameters
print(grid_search.best_params_)
```

```python
# create a decision tree classifier object with the best hyperparameters
model_rft = RandomForestClassifier(criterion=grid_search.best_params_['criterion'], 
                                   max_depth=grid_search.best_params_['max_depth'], random_state=3)

# train the classifier on the training set
model_rft.fit(X_train, y_train)

# predict the labels of the test set
y_pred = model_rft.predict(X_test)

print(f'The recall_score is : {recall_score(y_test, y_pred)}')  
print(f'The f1_score is : {f1_score(y_test, y_pred)}') 
print(f'The precision_score is : {precision_score(y_test, y_pred)}')  
print(f'The accuracy_score is : {accuracy_score(y_test, y_pred)}\n') 

scores = cross_val_score(model_rft, X_train, y_train, cv=5)
mean_accuracy = scores.mean()

print(f"Mean training accuracy (5-fold CV): {mean_accuracy:.2f}")
```
